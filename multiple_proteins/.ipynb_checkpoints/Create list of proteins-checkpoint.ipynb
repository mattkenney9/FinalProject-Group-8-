```python
#Create function that pulls data from molecule query and creates a DataFrame
def mol_query_to_list(molecule):
    return pd.DataFrame([[molecule['molecule_chembl_id'],
                        molecule['canonical_smiles'],
                        molecule['standard_value'],
                        molecule['assay_description'],
                        molecule['assay_type']]],
                        columns = ['molecule_chembl_id','canonical_smiles','standard_value','assay_description','assay_type'])
```

```python
#Iterates through the list of proteins, queries chembl for the list of molecules (and associated data) for each protein, 
#appends list of molecules to larger list
mol_list_of_lists = []
for target in target_df.iterrows():
    #Iterated target
    selected_target = target[1]['target_chembl_id']
    
    #Query
    activity = new_client.activity
    mol_query = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
    
    #Pull select query data to list 
    mol_list = list(map(mol_query_to_list, mol_query))
    
    #Append
    mol_list_of_lists.append(mol_list)
```

```python
#Standardization methods/code from Chanin lab
#https://github.com/chaninlab/estrogen-receptor-alpha-qsar/blob/master/02_ER_alpha_RO5.ipynb
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)
        
    return x
```

```python
#Converts standard value (IC50) to standardized pIC50
import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)
        
    return x
```
